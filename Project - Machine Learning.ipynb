```python
#2.2) Remove all the stopwords from the three speeches. Show the word count before and after the removal of stopwords. Show a sample sentence after the removal of stopwords.
```

```python
from nltk.corpus import stopwords
```

```python
roosevelt_words = nltk.word_tokenize(roosevelt_text)
kennedy_words = nltk.word_tokenize(kennedy_text)
nixon_words = nltk.word_tokenize(nixon_text)
```

```python
# Get the stopwords
stop_words = set(stopwords.words('english'))
```

```python
# Remove stopwords from each speech
roosevelt_words_without_stopwords = [word for word in roosevelt_words if word.lower() not in stop_words]
kennedy_words_without_stopwords = [word for word in kennedy_words if word.lower() not in stop_words]
nixon_words_without_stopwords = [word for word in nixon_words if word.lower() not in stop_words]
```

```python
# Calculate the word count before and after stopwords removal
roosevelt_word_count_before = len(roosevelt_words)
kennedy_word_count_before = len(kennedy_words)
nixon_word_count_before = len(nixon_words)
```

```python
roosevelt_word_count_after = len(roosevelt_words_without_stopwords)
kennedy_word_count_after = len(kennedy_words_without_stopwords)
nixon_word_count_after = len(nixon_words_without_stopwords)
```

```python
# Sample sentence after stopwords removal
sample_sentence = " ".join(roosevelt_words_without_stopwords[:20])  # Take the first 20 words
```

```python
# Print the results
print("Word count before removing stopwords:")
print("1941-Roosevelt.txt:", roosevelt_word_count_before)
print("1961-Kennedy.txt:", kennedy_word_count_before)
print("1973-Nixon.txt:", nixon_word_count_before)
print()

print("Word count after removing stopwords:")
print("1941-Roosevelt.txt:", roosevelt_word_count_after)
print("1961-Kennedy.txt:", kennedy_word_count_after)
print("1973-Nixon.txt:", nixon_word_count_after)
print()

print("Sample sentence after removing stopwords:")
print(sample_sentence)
```

```python
from collections import Counter
```

```python
roosevelt_words_without_stopwords = [word.lower() for word in roosevelt_words if word.lower() not in stop_words]
kennedy_words_without_stopwords = [word.lower() for word in kennedy_words if word.lower() not in stop_words]
nixon_words_without_stopwords = [word.lower() for word in nixon_words if word.lower() not in stop_words]
```

```python
roosevelt_word_counts = Counter(roosevelt_words_without_stopwords)
kennedy_word_counts = Counter(kennedy_words_without_stopwords)
nixon_word_counts = Counter(nixon_words_without_stopwords)
```

```python
top_three_roosevelt_words = roosevelt_word_counts.most_common(3)
top_three_kennedy_words = kennedy_word_counts.most_common(3)
top_three_nixon_words = nixon_word_counts.most_common(3)
```

```python
print("Top three words in Roosevelt's inaugural address:")
for word, count in top_three_roosevelt_words:
    print(f"{word}: {count}")

print("\nTop three words in Kennedy's inaugural address:")
for word, count in top_three_kennedy_words:
    print(f"{word}: {count}")

print("\nTop three words in Nixon's inaugural address:")
for word, count in top_three_nixon_words:
    print(f"{word}: {count}")
```
